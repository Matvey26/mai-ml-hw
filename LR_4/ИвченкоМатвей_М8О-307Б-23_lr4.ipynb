Выпишем признаки с большой зависимостью

````python
high_pairs = []
for c1, c2 in itertools.combinations(cols, 2):
    mi = mutual_info_score(df_codes[c1], df_codes[c2])
    if mi > 0.3:
        high_pairs.append((c1, c2, mi))

for col1, col2, mi in sorted(high_pairs, key=lambda t: -t[2])[:20]:
    print(col1, col2, round(mi, 2), sep=', ')
````

### 3. Построение Байесовсой сети

Данный метод представляет собой конструктивное построение структуры дискретной байесовской сети путем явного задания направленных ребер между узлами. В этом подходе структура сети создается декларативно - мы напрямую указываем список кортежей (родительский узел, дочерний узел), что определяет направленные зависимости между переменными.

````python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator

# Определяем структуру сети
model = DiscreteBayesianNetwork([
    # Класс зависит от самых информативных признаков
    ('odor', 'class'),
    ('spore-print-color', 'class'),
    ('gill-color', 'class'),
    ('ring-type', 'class'),
    
    # Блок "ножка"
    ('stalk-root', 'spore-print-color'),
    ('stalk-root', 'gill-color'),
    ('stalk-color-above-ring', 'stalk-color-below-ring'),
    ('stalk-root', 'stalk-color-above-ring'),
    ('stalk-root', 'stalk-color-below-ring'),
    
    # Блок "экология"
    ('population', 'habitat')
])

# Проверяем модель
print("Nodes:", model.nodes())
print("Edges:", model.edges())

# Обучаем параметры сети методом максимального правдоподобия
model.fit(df, estimator=MaximumLikelihoodEstimator)

# Пример: посмотреть CPD для 'class'
from pgmpy.inference import VariableElimination

infer = VariableElimination(model)
cpd_class = model.get_cpds('class')
from collections import Counter
for freq, cnt in Counter(cpd_class.values.ravel()).items():
    print(f'element: {freq:.3f}, frequency: {cnt}')
````

Получили очень разреженную модель. Давайте использовать Bayesian Estimator

````python
from pgmpy.estimators import BayesianEstimator

model.fit(df, estimator=BayesianEstimator, prior_type='BDeu', equivalent_sample_size=10)
````

````python
infer = VariableElimination(model)

cpd_class = model.get_cpds('class')

for freq, cnt in Counter(cpd_class.values.ravel()).items():
    print(f'element: {freq:.3f}, frequency: {cnt}')
````

Тогда уменьшим размер сети

````python
model = DiscreteBayesianNetwork([
    # class
    ('odor', 'class'),
    ('spore-print-color', 'class'),
    ('gill-color', 'class'),
    
    # сильные связи
    ('gill-color', 'spore-print-color'),
    ('odor', 'ring-type'),
    ('stalk-root', 'spore-print-color'),
    ('stalk-root', 'ring-type')
])
````

````python
model.fit(df, estimator=BayesianEstimator, prior_type='BDeu', equivalent_sample_size=25)

infer = VariableElimination(model)

cpd_class = model.get_cpds('class')
from collections import Counter
for freq, cnt in Counter(cpd_class.values.ravel()).items():
    print(f'element: {freq:.4f}, frequency: {cnt}')
````

Дело в том, что по своей структуре байесовские сети не интерполируют/экстраполируют данных, а модели по типу нейронных сетей - прекрасно с этим справляются

### 3. Получение вероятностей

Мы хотим получить вероятностные оценки переходов из одного состояние в другое. Для этого необходимо Байесовскую сеть подогнать по данные и получить таблички (CPT).

Conditional Probability Table (CPT) [Таблица Условных Вероятностей (CPT)] — это ключевой элемент, определяющий параметры Байесовской сети.

* Что это: Это таблица, которая содержит вероятность каждого возможного состояния узла (переменной) при условии всех возможных комбинаций состояний его родительских узлов (тех, от которых к нему идут стрелки/ребра).
* Роль: CPTs количественно определяют зависимости и силу связей между переменными, позволяя сети выполнять логический вывод и прогнозирование.
* Простой пример: В CPT для узла "Кашель", родительским узлом которого является "Грипп", будет указана вероятность кашля, если грипп есть (например, 90%), и вероятность кашля, если гриппа нет (например, 5%).

Чтобы получить значения вероятности переходов из разных узлов, нужно оценить эти вероятности. `HillClimbSearch` ищет только структуру, а `MaximumLikelihoodEstimator` и `BayesianEstimator` позволяют оценить вероятность.

Просмотр таблиц для конкретных узлов

````python
for _, node in model.edges():
    cpt = model.get_cpds(node)
    print(f"CPT for {node}:\n{cpt}")
````

### 6. Сравнение результатов с baseline-моделью (например, наивный байесовский классификатор из scikit-learn) по метрикам accuracy или log-likelihood.

````python
import numpy as np

# Список признаков, используемых в модели
features = ['stalk-root', 'odor', 'spore-print-color',
            'gill-color', 'ring-type']

# Считаем предсказания модели
y_true = (df['class'] == 'e').astype(int).values
y_pred = []

log_likelihood = 0.0

for idx, row in df.iterrows():
    evidence = row[features].to_dict()
    q = infer.query(variables=['class'], evidence=evidence)
    
    # Выбираем класс с максимальной вероятностью
    pred_class = q.values.argmax()  # если class закодирован как 0/1
    y_pred.append(pred_class)
    
    # Добавляем log-likelihood
    # pgmpy VariableElimination возвращает массив, где 0 индекс -> class=0, 1 -> class=1
    true_class = int(row['class'] == 'e')
    prob_true_class = q.values[true_class]
    log_likelihood += np.log(prob_true_class + 1e-12)  # добавляем eps, чтобы не было log(0)

# Accuracy
y_pred = np.array(y_pred)
accuracy = (y_pred == y_true).mean()

print(f"Accuracy: {accuracy:.4f}")
print(f"Log-Likelihood: {log_likelihood:.4f}")
````

Ну и посчитаем метрики для Naive Bayes

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, log_loss

y = df["class"]
X = df.drop(columns=["class"])

# Преобразуем категориальные признаки в числовые
X_encoded = X.apply(LabelEncoder().fit_transform)
y_encoded = LabelEncoder().fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Обучаем Наивный Байес
nb = CategoricalNB()
nb.fit(X_train, y_train)

# Предсказания
y_pred = nb.predict(X_test)
y_proba = nb.predict_proba(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

# Log-likelihood (log loss)
ll = -log_loss(y_test, y_proba, labels=nb.classes_, normalize=False)
print("Log-likelihood:", ll)
````
